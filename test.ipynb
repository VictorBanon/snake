```python
def addcookies(x,y,value,dic={}):
    
    if dic.__contains__(x):
        dic[x][y] = value
    else:
        dic[x]    = {y:value}
        
    return dic

a={} 
a[1]={0:"cookie"}  
a[1][1]="cookie"
a[1][2]="cookie"
a[2]={0:"cookie"} 
a[3]={1:"cookie"}
a[1][2]="cookie"

b={} 
b=addcookies(1,0,"cookie",b)
b=addcookies(1,1,"cookie",b)
b=addcookies(1,2,"cookie",b)
b=addcookies(2,0,"cookie",b)
b=addcookies(3,1,"cookie",b)
b=addcookies(1,2,"cookie",b)

print(a,b,sep="\n")
```
